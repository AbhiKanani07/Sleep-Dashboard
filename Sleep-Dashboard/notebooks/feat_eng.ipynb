## Feature List

| Feature | Description |
|--------|-------------|
| HOURS_OF_SLEEP_HOURS | Total sleep duration in hours |
| REM_SLEEP | % of REM sleep |
| DEEP_SLEEP | % of deep sleep |
| HEART_RATE_UNDER_RESTING | HR below resting % |
| bedtime_hour | Bedtime converted to float 0–24 |
| wakeup_hour | Wakeup converted to float 0–24 |
| weekday | Name of the day |

```python
# CELL 1: load data
import pandas as pd

fitbit = pd.read_csv("../data/clean/fitbit_clean.csv", parse_dates=["DATE"])
apple = pd.read_csv("../data/clean/apple_sleep_nightly_summary.csv", parse_dates=["DATE"])
```

```python
# CELL 2: implement time-variable features
fitbit["weekday"] = fitbit["DATE"].dt.day_name()
fitbit["week"] = fitbit["DATE"].dt.isocalendar().week
fitbit["month"] = fitbit["DATE"].dt.month
```

```python
# CELL 3: fitbit + apple health sleep logs -> numeric data conversion
fitbit["bedtime_dt"] = pd.to_datetime(fitbit["DATE"].dt.date.astype(str) + " " + fitbit["BEDTIME"])
fitbit["wakeup_dt"] = pd.to_datetime(fitbit["DATE"].dt.date.astype(str) + " " + fitbit["WAKEUP"])
```

```python
# CELL 4: display midpoint
fitbit["sleep_midpoint"] = fitbit["bedtime_dt"] + fitbit["HOURS_OF_SLEEP_HOURS"]*0.5*pd.Timedelta(hours=1)
```

```python
# CELL 5: create model variables

# Sleep Score (regression)
y = fitbit["SLEEP_SCORE"]

# Or good vs bad sleep (classification)
fitbit["good_sleep"] = (fitbit["HOURS_OF_SLEEP_HOURS"] >= 7).astype(int)
```
